```python
# Import necessry libraries and dependancies
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# target search example e.g. corona virus 
target = new_client.target
target_query = target.search('cyclooxygenase-2')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Select and retrieve bioactivity data 
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
# Filter for ic50 values only discarding other standard types for uniformity in data 
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type = 'IC50') # IC50 values in units: nano molar 
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.to_csv('cyclooxygenase2_raw_data.csv', index=False)
```

# Handling missing Data
#### Drop any rows(compounds) that have missing values for canonical smiles and standard type columns

```python
df2 = df[df.standard_value.notna() & df.canonical_smiles.notna()]
df2
```

```python
len(df.canonical_smiles.unique())
```

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

# Data preprocessing 
### combine 3 columns and bioactivity class into a dataframe

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

```python
df3.to_csv('cox2_preprocessed_01.csv', index=False)
```

# Labelling compounds as either active, inactive or intermediate

```python
df4 = pd.read_csv('cox2_preprocessed_01.csv')
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
    if i >= 10000:
        bioactivity_threshold.append('inactive')
    elif i <= 1000:
        bioactivity_threshold.append('active')
    else:
        bioactivity_threshold.append('intermediate')
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
df5.to_csv('cox2_data_curated_03.csv', index=False)
```
